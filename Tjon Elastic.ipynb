Code Exacted From MCEEP and Redone In Python by Douglas Higinbotham

Relativistic description of electron scattering on the deuteron
E. Hummel and J. A. Tjon

https://journals.aps.org/prc/abstract/10.1103/PhysRevC.49.21

```python
import numpy as np
import matplotlib.pyplot as plt

# Load The Hummel and Tjon Parameterizations

ia=np.loadtxt("ia.fc")
iamec=np.loadtxt("iamec.fc")
rsc=np.loadtxt("rsc.fc")
rscmec=np.loadtxt("rscmec.fc")


# Interpolate the Hummel and Tjon Parameterizations

from scipy.interpolate import interp1d
int_ia=interp1d(ia[:,0],ia[:,1],kind='cubic')
int_iamec=interp1d(iamec[:,0],iamec[:,1],kind='cubic')
int_rsc=interp1d(rsc[:,0],rsc[:,1],kind='cubic')
int_rscmec=interp1d(rscmec[:,0],rscmec[:,1],kind='cubic')
```

```python
# Plot the Results

myq2=np.linspace(0,20,1000)

plt.figure(dpi=150)
plt.subplot(2,2,1)
plt.plot(ia[:,0],ia[:,1],'o',label="IA",color='orange')
plt.plot(myq2,int_ia(myq2),'-',label="Interpolated IA",color='orange')
plt.xlabel('Q${^2}$ [fm$^{-2}$]',fontsize=10)
plt.ylabel('G$_C$',fontsize=10)
plt.legend(frameon=False)
plt.xlim(0,20)
plt.ylim(0)

plt.subplot(2,2,2)
plt.plot(iamec[:,0],iamec[:,1],'o',label="IAMEC",color='green')
plt.plot(myq2,int_iamec(myq2),'-',label="IAMEC",color='green')
plt.xlabel('Q${^2}$ [fm$^{-2}$]',fontsize=10)
plt.ylabel('G$_C$',fontsize=10)
plt.legend(frameon=False)
plt.xlim(0,20)
plt.ylim(0)

plt.subplot(2,2,3)
plt.plot(rsc[:,0],rsc[:,1],'o',label="RSC",color='blue')
plt.plot(myq2,int_rsc(myq2),'-',label="Interpolated RSC",color='blue')
plt.xlabel('Q${^2}$ [fm$^{-2}$]',fontsize=10)
plt.ylabel('G$_C$',fontsize=10)
plt.legend(frameon=False)
plt.xlim(0,20)
plt.ylim(0)

plt.subplot(2,2,4)
plt.plot(rscmec[:,0],rscmec[:,1],'o',label="RSCMEC",color='red')
plt.plot(myq2,int_rscmec(myq2),'-',label="RSCMEC",color='red')
plt.xlabel('Q${^2}$ [fm$^{-2}$]',fontsize=10)
plt.ylabel('G$_C$',fontsize=10)
plt.legend(frameon=False)
plt.xlim(0,20)
plt.ylim(0)

plt.tight_layout()
plt.savefig('HummelAndTjon.png')
plt.show()
```

EXPLANATION OF DEUTERON ELASTIC FORM FACTOR FILES

ia       (relativistic IA)

iamec    (relativistic IA + MEC)

rsc      (relativistic IA       w/ Reid Soft Core)

rscmec   (relativistic IA + MEC w/ Reid Soft Core)
 
IA : Hohler form factors
MEC: (pi-rho-gamma plus omega-sigma-gamma contribution)
No Lorentz boosts and negative spinor states.
